## Great Expectations
### 1. No null values

```python
validator.expect_column_values_to_not_be_null('brand')
```

### 2. Unique Values for column id

```python
validator.expect_column_values_to_be_unique('id')
```

### 3. Expect values to be between 0 and 6 million for column price

```python
validator.expect_column_values_to_be_between(
    column='price', min_value=0, max_value=6000000
)
```

### 4. Expect values of 1 and 0 to be in a set of column trusted_seller

```python
allowed_values = [0,1]
validator.expect_column_values_to_be_in_set(
    column='trusted_seller',
    value_set=allowed_values
)
```

### 5. Expect column sold to be either an integer or a float type data

```python
validator.expect_column_values_to_be_in_type_list('sold', ['integer', 'float'])
```

### 6. Expect column year_produced values to be higher than column punctual

```python
validator.expect_column_pair_values_a_to_be_greater_than_b(
    column_A='year_produced',
    column_B='punctual'
)
```

### 7. Expect the most common value for column movement to be automatic

```python
validator.expect_column_most_common_value_to_be_in_set(
    column='movement',
    value_set=['Automatic']
)
```

### 8. Expect the median value of column review to be between 0 and 300

```python
validator.expect_column_median_to_be_between('review', 0, 300)
```
